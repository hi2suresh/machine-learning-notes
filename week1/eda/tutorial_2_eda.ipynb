# üìä What is Exploratory Data Analysis (EDA)?
Exploratory Data Analysis (EDA) is the critical first step in any data science project where you investigate, summarize, and visualize your dataset to understand its main characteristics before applying machine learning algorithms.
Think of EDA as detective work - you're gathering clues about your data to understand what story it tells and what patterns might be hiding within it.

üéØ Why is EDA Essential?
1. Data Quality Assessment

- Identify missing values, outliers, and inconsistencies
- Detect data entry errors and anomalies
- Understand data completeness and reliability

2. Understanding Data Structure

- Learn about variable types (numerical, categorical, ordinal)
- Understand relationships between variables
- Identify data distribution patterns

3. Feature Engineering Insights

- Discover which variables might be most predictive
- Identify potential feature transformations needed
- Spot opportunities for creating new features

4. Model Selection Guidance

- Understanding data distributions helps choose appropriate algorithms
- Identify if data meets assumptions of certain models
- Determine preprocessing steps needed

5. Business Understanding

- Generate hypotheses about the data
- Validate or challenge domain assumptions
C- ommunicate findings to stakeholders

Remember: "Garbage in, garbage out" - EDA helps ensure you're not feeding poor-quality data into your models!

````python
# Convert to categorical data types for better analysis and memory efficiency

# Why we do this:
# - Ordered categories (ordinal) preserve natural ranking relationships
# - Unordered categories (nominal) are more memory efficient than strings
# - Pandas can perform category-specific operations and optimizations
# - Prevents incorrect numerical operations on categorical data

# Convert to ordered categorical (ordinal) - these have natural ranking
df['experience_level'] = pd.Categorical(df['experience_level'], 
                                       categories=['EN', 'MI', 'SE', 'EX'], 
                                       ordered=True)

df['company_size'] = pd.Categorical(df['company_size'], 
                                   categories=['S', 'M', 'L'], 
                                   ordered=True)

df['remote_ratio'] = pd.Categorical(df['remote_ratio'], 
                                   categories=[0, 50, 100], 
                                   ordered=True)

# Convert to unordered categorical (nominal) - no natural ranking
df['employment_type'] = pd.Categorical(df['employment_type'], ordered=False)

# Verify the changes
print("Data types after conversion:")
print(df[['experience_level', 'company_size', 'remote_ratio', 'employment_type']].dtypes)

print("\nCategory information:")
print(f"Experience level categories: {df['experience_level'].cat.categories.tolist()}")
print(f"Company size categories: {df['company_size'].cat.categories.tolist()}")
print(f"Remote ratio categories: {df['remote_ratio'].cat.categories.tolist()}")
print(f"Employment type categories: {df['employment_type'].cat.categories.tolist()}")

# After converting to categorical, reindex to set the proper order

# Reorder the categories to ensure proper plotting order
df['experience_level'] = df['experience_level'].cat.reorder_categories(['EN', 'MI', 'SE', 'EX'])
df['company_size'] = df['company_size'].cat.reorder_categories(['S', 'M', 'L']) 
df['remote_ratio'] = df['remote_ratio'].cat.reorder_categories([0, 50, 100])

# Now all plots will automatically use the correct order!
````

## 5.  Comparing Measurements 

````python
## Task: Perform the following comparisons to understand salary patterns:

# 1. Compare salary distributions by experience level
#    - Create side-by-side boxplots for each experience level
#    - Calculate mean and median salary for each level

# 1. Compare salary distributions by experience level
print("="*60)
print("1. SALARY COMPARISON BY EXPERIENCE LEVEL")
print("="*60)


company_size_salary = (df.groupby('company_size')['salary_in_usd']
                      .agg(
                          mean_salary='mean',
                          median_salary='median',
                          count='count'
                      )
                      .reset_index()
                      .sort_values('mean_salary', ascending=False)
                      .round(2))

print("Salary statistics by company size:")
print(company_size_salary)

# Visual comparison
plt.figure(figsize=(10, 6))
df.boxplot(column='salary_in_usd', by='company_size', ax=plt.gca())
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.suptitle('')
plt.show()

# 3. Compare remote vs non-remote work salaries
print("\n" + "="*60)
print("3. SALARY COMPARISON BY REMOTE WORK")
print("="*60)

remote_salary_stats = (df.groupby('remote_ratio')['salary_in_usd']
                      .agg(
                          mean_salary='mean',
                          median_salary='median',
                          count='count'
                      )
                      .reset_index()
                      .sort_values('mean_salary', ascending=False)
                      .round(2))

print("Salary statistics by remote work ratio:")
print(remote_salary_stats)

# Add readable labels
remote_labels = {0: 'No Remote\n(0%)', 50: 'Hybrid\n(50%)', 100: 'Full Remote\n(100%)'}
df_plot = df.copy()
df_plot['remote_label'] = df_plot['remote_ratio'].map(remote_labels)

plt.figure(figsize=(10, 6))
df_plot.boxplot(column='salary_in_usd', by='remote_label', ax=plt.gca())
plt.title('Salary Distribution by Remote Work Policy')
plt.xlabel('Remote Work Policy')
plt.ylabel('Salary (USD)')
plt.suptitle('')
plt.show()
#TODO: 
# 2. Compare salary by company size
#    - Show salary distribution across small, medium, and large companies
#    - Identify which company size pays the most

# 3. Compare remote vs non-remote work salaries
#    - Group by remote_ratio (0%, 50%, 100%)
#    - Compare salary statistics

````

````python
# 4. Compare salary trends over years
#    - Show how average salaries changed from 2020-2023
#    - Identify salary growth patterns

print("\n" + "="*60)
print("4. SALARY TRENDS OVER YEARS")
print("="*60)

yearly_salary_trends = (df.groupby('work_year')['salary_in_usd']
                       .agg(
                           mean_salary='mean',
                           median_salary='median',
                           count='count'
                       )
                       .reset_index()
                       .round(2))

print("Salary trends by year:")
print(yearly_salary_trends)

# Calculate year-over-year growth
yearly_salary_trends['salary_growth'] = yearly_salary_trends['mean_salary'].pct_change() * 100
print("\nYear-over-year salary growth:")
print(yearly_salary_trends[['work_year', 'mean_salary', 'salary_growth']].round(2))

# Visual trend analysis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Line plot for trends
ax1.plot(yearly_salary_trends['work_year'], yearly_salary_trends['mean_salary'], 
         marker='o', linewidth=2, markersize=8, color='blue')
ax1.set_title('Average Salary Trend Over Years')
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Salary (USD)')
ax1.grid(True, alpha=0.3)

# Boxplot by year
df.boxplot(column='salary_in_usd', by='work_year', ax=ax2)
ax2.set_title('Salary Distribution by Year')
ax2.set_xlabel('Year')
ax2.set_ylabel('Salary (USD)')

plt.tight_layout()
plt.show()

# Summary insights
print("\n" + "="*60)
print("KEY COMPARISON INSIGHTS")
print("="*60)

# Experience level insights
highest_exp_salary = experience_salary_stats.loc[experience_salary_stats['mean_salary'].idxmax()]
print(f"üí∞ Highest paying experience level: {highest_exp_salary['experience_level']} (${highest_exp_salary['mean_salary']:,.0f})")

# Company size insights
highest_company_salary = company_size_salary.iloc[0]  # Already sorted
print(f"üè¢ Highest paying company size: {highest_company_salary['company_size']} (${highest_company_salary['mean_salary']:,.0f})")

# Remote work insights
highest_remote_salary = remote_salary_stats.iloc[0]  # Already sorted
remote_label = {0: 'No Remote', 50: 'Hybrid', 100: 'Full Remote'}[highest_remote_salary['remote_ratio']]
print(f"üè† Highest paying remote policy: {remote_label} (${highest_remote_salary['mean_salary']:,.0f})")

# Yearly growth
latest_growth = yearly_salary_trends['salary_growth'].iloc[-1]
if pd.notna(latest_growth):
    print(f"üìà Latest year salary growth: {latest_growth:.1f}%")
````

````python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Select variables to analyze
vars_to_plot = ['employee_residence', 'company_location', 'job_title']

for i, var in enumerate(vars_to_plot):
    # Get top 10 and filter data
    top_10 = df[var].value_counts().head(10).index
    data_subset = df[df[var].isin(top_10)]
    
    # Create boxplot
    sns.boxplot(data=data_subset, x=var, y='salary_in_usd', ax=axes[i])
    axes[i].set_title(f'Top 10 {var.replace("_", " ").title()}')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
````

**Observations** : 
- Organizations headquartered in the United States consistently offer the most competitive compensation packages compared to companies in other regions. This suggests that the U.S. remains the premium market for AI/ML talent globally.
- Employee Residence Premium: Data science professionals residing in the United States command significantly higher salaries than their counterparts in other countries, regardless of company location. 
- Limited Job Title Correlation: The analysis reveals a surprisingly weak correlation between specific job titles and salary levels. This finding suggests that factors such as experience level, company characteristics, and geographic location are more influential determinants of compensation than the particular job title held.
